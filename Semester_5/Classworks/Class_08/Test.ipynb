```python
n = int(input("Enter a five digit number"))
divisor = 10000
while n > 0:
    print(n // divisor, end = " ")
    n %= divisor
    divisor //= 10
```

```python
total_gallons, total_miles = 0, 0
while True:
    gallons = float(input("Enter the gallons used: "))
    if gallons == -1:
        break;
    miles = float(input("Enter the miles driven: "))
    total_gallons, total_miles = total_gallons + gallons, total_miles + miles
    print(f"The miles/gallon for this tank was: {(miles / gallons):.2f}")

print(f"The overall average miles/gallon was: {(total_miles / total_gallons):.2f}")
```

```python
num = int(input("Enter a number: "))
num2 = num
check_num = 0
while num > 0:
    digit = num % 10
    check_num = (check_num * 10) + digit
    num //= 10

if num2 == check_num:
    print(f"{num2} is a palindrome number.")
else :
    print(f"{num2} is not a palindrome number.")
```

```python
num = input("Enter a number: ")
i, j, isPalindrome = 0, len(num) - 1, True
while i <= j:
    if num[i] != num[j]:
        isPalindrome = False
    i = i + 1
    j = j - 1
if isPalindrome:
    print(f"{num} is a palindrome number.")
else :
    print(f"{num} is not a palindrome number.")
```

```python
pi, denominator = 4, 3
count, sign_count = 1, 1
pi_314, pi_3141, pi_31415, pi_314159 = None, None, None, None
print("Count            PI")
print(f"1                {pi}")
while count < 1000:
    if sign_count == 1:
        pi -= (4 / denominator)
        denominator = denominator + 2
    else:
        pi += (4 / denominator)
        denominator = denominator + 2

    if pi_314 == None and f"{pi:.2f}" == '3.14':
        pi_314 = 'got target'
        print(f"{count} terms are required to get 3.14 for the first time.")

    if pi_3141 == None and f"{pi:.3f}" == '3.141':
        pi_3141 = 'got target'
        print(f"{count} terms are required to get 3.141 for the first time.")

    if pi_31415 == None and f"{pi:.4f}" == '3.1415':
        pi_31415 = 'got target'
        print(f"{count} terms are required to get 3.1415 for the first time.")

    if pi_314159 == None and f"{pi:.5f}" == '3.14159':
        pi_314159 = 'got target'
        print(f"{count} terms are required to get 3.14159 for the first time.")
        
    sign_count *= -1
    count = count + 1
        
    print(f"{count:}         {pi}")
```
