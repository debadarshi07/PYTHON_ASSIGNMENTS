```python
import pandas as pd
import numpy as np
```

```python
# (a) Series from list [7, 11, 13, 17]
series_a = pd.Series([7, 11, 13, 17])
print(f"(a): Series from list [7, 11, 13, 17]:\n{series_a}\n")

# (b) Series with five elements of 100.0
# series_b = pd.Series([100.0] * 5)
series_b = pd.Series(100.0, range(5))
print(f"(b): Series with five elements of 100.0:\n{series_b}\n")

# (c) Series with 20 random numbers in the range 0 to 100
series_c = pd.Series(np.random.randint(0, 101, 20))
print(f"(c): Series with 20 random numbers:\n{series_c}\n")
print(f"Descriptive statistics:\n{series_c.describe()}\n")

# (d)
temperatures = pd.Series( [98.6, 98.9, 100.2, 97.9], index=["Julie", "Charlie", "Sam", "Andrea"] )
print(f"(d): Temperatures Series:\n{temperatures}\n")

# (e)
temp_dict = temperatures.to_dict()
new_series = pd.Series(temp_dict)
print(f"(e): Dictionary formed from temperatures series:\n{temp_dict}\n")
print(f"Series initialized from dictionary:\n{new_series}\n")
```
