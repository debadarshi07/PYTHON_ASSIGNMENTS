```python
import pandas as pd
```

```python
# Series using pandas

# different ways to define series

series1 = pd.Series([5,42,8,11,3,5,42])
print(f"series1: \n{series1}\n\n")

# for custom indexing we need to use dictionary
series2 = pd.Series({"one": 18, "two": 22, "three": 25, "four": 26, "five": 37, "six": 3333, "seven": 87, "eight": 56})
print(f"series2: \n{series2}\n\n")

series3 = pd.Series(series1)
print(f"series3: \n{series3}\n\n")

series4 = pd.Series(7)
print(f"series4: \n{series4}\n\n")

series5 = pd.Series(23, range(10))
print(f"series5: \n{series5}\n\n")

series6 = pd.Series(55, ['a', 'b', 'c', 'd'])   # the second argument should be an iterable with the indices values
print(f"series6: \n{series6}\n\n")

series7 = pd.Series([10, 20, 30, 40, 50], ['ten', 'twenty', 'thirty', 'forty', 'fifty'])
print(f"series7: \n{series7}\n")
```

```python
# Accessing and Updating series elements
grades = pd.Series([79, 90, 85, 75, 55, 82, 66], ['a', 'b', 'c', 'd', 'e', 'f', 'g'])

print(f"grades.iloc[0]: {grades.iloc[0]}\ngrades.iloc[1]: {grades.iloc[1]}\n")
print(f"grades['c']: {grades['c']}\ngrades['d']: {grades['d']}\n")
print(f"grades.e: {grades.e}\ngrades.f: {grades.f}\n")

grades.c = 69
print(f"grades: \n{grades}\n")
print(f"grades.dtype: {grades.dtype}")
```

```python
# Descriptive stats of a series elements

print(f"grades.count(): {grades.count()}")
print(f"grades.mean(): {grades.mean()}")
print(f"grades.std(): {grades.std()}")
print(f"grades.var(): {grades.var()}")
print(f"grades.min(): {grades.min()}")
print(f"grades.max(): {grades.max()}")
```

```python
import numpy as np
ser = pd.Series(np.arange(11))
print(ser)
```

```python
# Series of strings
hardware = pd.Series(['Hammer', 'Saw', 'Wrench', 'Pickaxe'])

print(f"hardware.str.contains('a'):\n{hardware.str.contains('a')}\n")

print(f"hardware.str.upper():\n{hardware.str.upper()}\n")

print(f"hardware.str.len():\n{hardware.str.len()}\n")
```

```python
integers = np.random.randint(60, 100, 5)
temperatures = pd.Series(integers)
print(f'temperatures: \n{temperatures}\n')
print(f"Lowest: {temperatures.min()}\nHighest: {temperatures.max()}\nAverage: {temperatures.mean()}\n")
print(f"Descriptive statistics: \n{temperatures.describe()}\n")
```

```python
# DataFrame
students = {'Name': ['Narendra', 'Rahul', 'Amit', 'Sonia', 'Nirmala'], 
           'Class': ['PM', 'LoO', 'HM', 'MP', 'FM'], 
           'Roll no': [1, 69, 2, 4, 8], 
           'CGPA': [4.00, 2, 9.5, 6, 8]}

students_df = pd.DataFrame(students, index = ['i', 'ii', 'iii', 'iv', 'v'])

print(f"Students DataFrame: \n\n{students_df}\n\n")
print(f"grades.index: {students_df.index}")
```

```python
# Accessing DataFrame elements

print(f"students_df['Name']: \n{students_df['Name']}\n")
print(f"students_df['Class']: \n{students_df['Class']}\n")

print(f"students_df.CGPA: \n{students_df.CGPA}\n")

print(f"students_df.loc['ii']: \n{students_df.loc['ii']}\n")  # Access data using the name of the custom index
print(f"students_df.iloc[1]: \n{students_df.iloc[1]}\n")  # Access data using the default 0 based indexing
```

```python
# Slicing
print(f"students_df[1 : 4]: \n{students_df[1 : 4]}\n")  # Slices using 0 based indexing
print(f"students_df.loc['ii' : 'iv']: \n{students_df['ii' : 'iv']}\n")  # Includes both the start and end custom indices
print(f"students_df.iloc[2 : 4]: \n{students_df.iloc[2 : 4]}")  # Same as normal index based slicing
```

```python
# Slicing with Mixed Slices and Lists
print(f"students_df.loc['ii' : 'iv', ['Name', 'CGPA']]: \n{students_df.loc['ii' : 'iv', ['Name', 'CGPA']]}\n")
print(f"students_df.iloc[[1, 3], 1 : 4]: \n{students_df.iloc[[1, 3], 1 : 4]}")
```

```python
# Boolean Indexing
print(f"students_df[students_df.CGPA > 5]:")
students_df[students_df.CGPA > 5]
```

```python
# Accessing and Updating a specific cell

print(f"students_df.at['iii', 'Roll no']: {students_df.at['iii', 'Roll no']}\n")
# Updating
students_df.at['iv', 'Class'] = 'Maa'
print(f"students_df: \n{students_df}\n\n")


# Integer based access and updation
print(f"students_df.iat[1, 2]: {students_df.iat[1, 2]}\n")
# Updating
students_df.iat[2, 2] = 30
print(f"students_df: \n{students_df}")
```
