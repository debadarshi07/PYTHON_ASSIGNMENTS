```python
# Editing a line in an existing file
with open('employees.txt', mode = 'r+') as emp:
    employees = emp.readlines()
    for i, employee in enumerate(employees):
        if employee.find('Akhilesh') != -1:
            emp_id, emp_name, emp_salary = employee.split()
            emp_name = 'Yogi'
            employee = f'{emp_id} {emp_name} {emp_salary}\n'
            employees[i] = employee
        else:
            pass
            
    emp.seek(0)
    emp.truncate(0)
    emp.writelines(employees)
```

```python
with open('employees.txt', mode = 'r+') as emp:
    employees = emp.readlines()
    emp.seek(0)
    for employee in employees:
        if employee.find('Yogi') != -1:
            emp_id, emp_name, emp_salary = employee.split()
            emp_name = 'Akhilesh'
            emp.write(f'{emp_id} {emp_name} {emp_salary}\n')
        else:
            emp.write(employee)
```

```python
# Handling CSV files

# Writing to a CSV file
import csv

with open('accounts.csv', mode = 'w', newline = '') as acc:
    writer = csv.writer(acc)
    writer.writerow(['Account Number', 'Account Holder Name', 'Amount(lakhs)'])
    writer.writerows([ ['1111', 'Name1', 25], ['1112', 'Name2', 43], ['1113', 'Name3', 33], ['1114', 'Name4', 75] ])
```

```python
# Reading a CSV file
with open('accounts.csv', mode = 'r', newline = '') as acc:
    reader = csv.reader(acc)

    for record in reader:
        acc_no, acc_name, acc_amt = record
        print(f'{acc_no: ^25} {acc_name: ^25} {acc_amt: ^25}')
```

```python
happiness_index = pd.read_csv(r'https://raw.githubusercontent.com/MainakRepositor/Datasets/refs/heads/master/World%20Happiness%20Data/2020.csv')
happiness_index
```

```python
all_regions = happiness_index['Regional indicator']
all_regions
```

```python
import matplotlib.pyplot as plt
from collections import Counter
```

```python
occurences = Counter(all_regions)
regions = sorted(set(all_regions))

x = regions
y = [occurences[region] for region in regions]

plt.figure(figsize = (10, 4))
plt.title('Country counts in different regions', fontdict = {'family': 'serif', 'color': 'red', 'size': 20})
plt.xlabel('Regions', fontdict = {'family': 'serif', 'color': 'blue', 'size': 20})
plt.ylabel('Counts', fontdict = {'family': 'serif', 'color': 'blue', 'size': 20})

plt.xticks(rotation = 90)
plt.bar(x, y, color = 'orange', edgecolor = 'black')

plt.show()
```

```python
ladder_score = happiness_index['Ladder score']
ladder_score = ladder_score.dropna()

plt.figure(figsize = (12, 8))
plt.title('Countries with happiness index', fontdict = {'family': 'serif', 'color': 'red', 'size': 20})
plt.xlabel('Ladder score', fontdict = {'family': 'serif', 'color': 'blue', 'size': 15})
plt.ylabel('Countries', fontdict = {'family': 'serif', 'color': 'blue', 'size': 15})

bins_edges = np.linspace(min(ladder_score), max(ladder_score), 16)
plt.xticks(bins_edges, rotation = 90)

plt.hist(ladder_score, color = 'maroon', edgecolor = 'black', bins = 15)

plt.show()
```
