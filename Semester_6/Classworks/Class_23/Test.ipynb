```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
digits.data.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2, random_state=10)
```

```python
r_data = pca.fit_transform(digits.data)
```

```python
r_data.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
axes = plt.scatter(r_data[:,0], r_data[:,1], c=digits.target, cmap=plt.get_cmap('nipy_spectral'))
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, random_state=10)
```

```python
r_data = tsne.fit_transform(digits.data)
```

```python
axes = plt.scatter(r_data[:,0], r_data[:,1], c=digits.target, cmap=plt.get_cmap('nipy_spectral'))
colorbar = plt.colorbar(axes)
```
